```python
con = duckdb.connect(database=':memory:', read_only=False)
```

```python
con.execute("""SELECT COUNT(*) FROM '/home/morrita/w/hn-full/*.parquet'""").df()
```

How can I see the schema in the Parquet file?

```python
con.execute("""CREATE VIEW hn AS SELECT * FROM '/home/morrita/w/hn-full/*.parquet'""").fetchall()
```

```python
con.execute("""PRAGMA table_info('hn')""").df()
```
